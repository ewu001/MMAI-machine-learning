```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, auc, roc_auc_score

import preprocess as preproc
import util as utility
import model as md
```

```python
train_data = pd.read_csv('data store/training_dataset.csv')
print(train_data.shape)
train_data.head()
```

```python
eval_data = pd.read_csv('data store/evaluation_dataset.csv')
print(eval_data.shape)
```

```python
#  Prepare preprocessor 
#  Go through transformation process and generate processed dataframe

features_dropna = ["Tobin's Q", "EPS", "Leverage Ratio", "Return on Equity", "Market Book Ratio", "Operational Margin", "Assets Growth", "Productivity", "Sales Growth", "Employee Growth"]
features_impute = ["Liquidity", "Profitability", "Asset Turnover"]
features_todrop = ["Unnamed: 0", "Liquidity"]
target = "BK"
train_data_object = preproc.preprocessor(train_data)
eval_data_object = preproc.preprocessor(eval_data)

processed_df = preproc.feature_preprocessor(train_data_object, features_dropna, features_impute, features_todrop, target)
processed_eval_df = preproc.feature_preprocessor(eval_data_object, features_dropna, features_impute, features_todrop, target)
```

```python
df_explore = utility.df_explore_helper(processed_df)
df_explore

processed_df = processed_df.rename(columns={"Tobin's Q": 'Tobin_Q'})
processed_df = processed_df.rename(columns={"Asset Turnover": 'Asset_Turnover'})
processed_df = processed_df.rename(columns={"Assets Growth": 'Assets_Growth'})
processed_df = processed_df.rename(columns={"Data Year - Fiscal": 'Fiscal_year'})
processed_df = processed_df.rename(columns={"Employee Growth": 'Employee_Growth'})
processed_df = processed_df.rename(columns={"Leverage Ratio": 'Leverage_Ratio'})
processed_df = processed_df.rename(columns={"Return on Equity": 'Return_Equity'})
processed_df = processed_df.rename(columns={"Market Book Ratio": 'Market_Book_Ratio'})
processed_df = processed_df.rename(columns={"Operational Margin": 'Operational_Margin'})
processed_df = processed_df.rename(columns={"Sales Growth": 'Sales_Growth'})

processed_eval_df = processed_eval_df.rename(columns={"Tobin's Q": "Tobin_Q"})
processed_eval_df = processed_eval_df.rename(columns={"Asset Turnover": 'Asset_Turnover'})
processed_eval_df = processed_eval_df.rename(columns={"Assets Growth": 'Assets_Growth'})
processed_eval_df = processed_eval_df.rename(columns={"Data Year - Fiscal": 'Fiscal_year'})
processed_eval_df = processed_eval_df.rename(columns={"Employee Growth": 'Employee_Growth'})
processed_eval_df = processed_eval_df.rename(columns={"Leverage Ratio": 'Leverage_Ratio'})
processed_eval_df = processed_eval_df.rename(columns={"Return on Equity": 'Return_Equity'})
processed_eval_df = processed_eval_df.rename(columns={"Market Book Ratio": 'Market_Book_Ratio'})
processed_eval_df = processed_eval_df.rename(columns={"Operational Margin": 'Operational_Margin'})
processed_eval_df = processed_eval_df.rename(columns={"Sales Growth": 'Sales_Growth'})

# write to csv
processed_df.to_csv("data store/training_dataset_processed.csv", encoding='utf-8', index=False)
processed_eval_df.to_csv("data store/evaluation_dataset_processed.csv", encoding='utf-8', index=False)

processed_df.to_csv("data store/training_dataset_processed_tf.csv", encoding='utf-8', index=False, header=False)
processed_eval_df.to_csv("data store/evaluation_dataset_processed_tf.csv", encoding='utf-8', index=False, header=False)

processed_df.head()
```
