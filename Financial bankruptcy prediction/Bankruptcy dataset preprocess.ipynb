```python
import pandas as pd

from helper import preprocess as preproc
from helper import util
from helper import ml_model as md
```

```python
train_data = pd.read_csv('data store/training_dataset.csv')
print(train_data.shape)
train_data.head()
```

```python
eval_data = pd.read_csv('data store/evaluation_dataset.csv')
print(eval_data.shape)
```

```python
#  Prepare preprocessor 
#  Go through data transformation process and generate processed dataframe

features_dropna = ["Tobin's Q", "EPS", "Leverage Ratio", "Return on Equity", "Market Book Ratio", "Operational Margin", "Assets Growth", "Productivity", "Sales Growth", "Employee Growth"]
features_impute = ["Liquidity", "Profitability", "Asset Turnover"]
features_todrop = ["Unnamed: 0", "Liquidity"]
target = "BK"
train_data_object = preproc.preprocessor(train_data)
eval_data_object = preproc.preprocessor(eval_data)

processed_df = preproc.feature_preprocessor(train_data_object, features_dropna, features_impute, features_todrop, target)
processed_eval_df = preproc.feature_preprocessor(eval_data_object, features_dropna, features_impute, features_todrop, target)
```

```python
processed_df = preproc.clean_column_name(processed_df)
processed_eval_df =  preproc.clean_column_name(processed_eval_df)

# write to csv
processed_df.to_csv("data store/training_dataset_processed.csv", encoding='utf-8', index=False)
processed_eval_df.to_csv("data store/evaluation_dataset_processed.csv", encoding='utf-8', index=False)

processed_df.to_csv("data store/training_dataset_processed_tf.csv", encoding='utf-8', index=False, header=False)
processed_eval_df.to_csv("data store/evaluation_dataset_processed_tf.csv", encoding='utf-8', index=False, header=False)

processed_df.head()
```
